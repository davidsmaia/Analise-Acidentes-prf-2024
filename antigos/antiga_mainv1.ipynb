# Importando e Preparando os Dados para a Análise

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados2024 = pd.read_csv("acidentes2024_todas_causas_tipos.csv", sep = ";", encoding= "latin-1")
pd.set_option("display.max_columns", None) # Permite a visualização de todas as colunas
display(dados2024)
```

```python
# Removendo colunas que não serão utilizadas

colunas_remover = ["id","pesid","id_veiculo","marca","ano_fabricacao_veiculo","latitude", "longitude","delegacia", "uop","regional"]

dados2024 = dados2024.drop(columns=colunas_remover)

display(dados2024)
```

```python
# Filtrando apenas os acidentes com vítimas fatais

dados_vitimas_fatais = dados2024[dados2024["classificacao_acidente"] == "Com Vítimas Fatais"]
display(dados_vitimas_fatais.head())
```

# Pergunta 1 - Quais os estados com o maior número de registros acidentes ?

```python
acidentes_por_estado = dados2024["uf"].value_counts()
# display(acidentes_por_estado)

# Gerar tabela formatada em Markdown
print("| Posição |   Estado   | Quantidade |")
print("|---------|------------|------------|")
for posicao, (estado, qtd) in enumerate(acidentes_por_estado.head(10).items()):

    if posicao < 9:
        print(f"| 0{posicao + 1}º     | {estado.ljust(9)}  | {str(qtd).rjust(9)}  |")
    
    else:
        print(f"| {posicao + 1}º     | {estado.ljust(9)}  | {str(qtd).rjust(9)}  |")
```

```python
# Plotando os 10 estados com mais acidentes registrados

plt.figure(figsize=(12,6))
acidentes_por_estado.head(10).plot(kind = "bar", color = "blue")
plt.title("Registros de Acidentes em Rodovias Federais - 10 Maiores Estados")
plt.xlabel("Estados (UF)")
plt.ylabel("Registros de Acidentes")
plt.xticks(rotation = 0)
plt.savefig("acidentes_por_estado")
plt.show()
```

# Pergunta 2 - Quais os estados com o maior número de acidentes com Vítimas Fatais ?

```python
acidentes_vitimas_fatais_por_estado = dados_vitimas_fatais["uf"].value_counts()
# display(acidentes_vitimas_fatais_por_estado)

# Gerar tabela formatada em Markdown
print("| Posição |   Estado   | Quantidade |")
print("|---------|------------|------------|")
for posicao, (estado, qtd) in enumerate(acidentes_vitimas_fatais_por_estado.head(10).items()):

    if posicao < 9:
        print(f"| 0{posicao + 1}º     | {estado.ljust(9)}  | {str(qtd).rjust(9)}  |")
    
    else:
        print(f"| {posicao + 1}º     | {estado.ljust(9)}  | {str(qtd).rjust(9)}  |")
```

```python
# Plotando o gráfico com os estados que possuem os maiores registros de Acidentes com Vítimas Fatais

plt.figure(figsize=(12,6))
acidentes_vitimas_fatais_por_estado.head(10).plot(kind = "bar", color = "red")
plt.title("Registros de Acidentes com Vítimas Fatais em Rodovias Federais - 10 Maiores Estados")
plt.xlabel("Estados (UF)")
plt.ylabel("Registros de Acidentes com Vítimas Fatais")
plt.xticks(rotation = 0)
plt.savefig("fatalidades_por_estado")
plt.show()
```

# Pergunta 3 - Quais os tipos de veículos mais se envolveram em acidentes ?

```python
# Separando os valores na variável
acidentes_por_veiculo = dados2024["tipo_veiculo"].value_counts()
display(acidentes_por_veiculo.head(5))

# Gerar tabela formatada em Markdown
print("| Posição |   Tipo de Veículo   | Quantidade |")
print("|---------|---------------------|------------|")
for posicao, (veiculo, qtd) in enumerate(acidentes_por_veiculo.head(10).items()):

    if posicao < 9:
        print(f"| 0{posicao + 1}º     | {veiculo.ljust(18)}  | {str(qtd).rjust(9)}  |")
    
    else:
        print(f"| {posicao + 1}º     | {veiculo.ljust(18)}  | {str(qtd).rjust(9)}  |")
```

```python
# Plotando o gráfico

plt.figure(figsize=(12,6))
acidentes_por_veiculo.head(10).sort_values(ascending=True).plot(kind = "barh", color = "blue")
plt.title("Veículos com Maiores Registro de Acidentes em Rodovias Federais")
plt.xlabel("Registros de Acidentes")
plt.xticks(rotation = 0)
plt.savefig("acidentes_por_veiculo")
plt.show()
```

# Pergunta 4 - Quais veículos possuem os maiores números de Acidentes com Vítimas Fatais ?

```python
# Separando os valores na variável
fatalidades_por_veiculo = dados_vitimas_fatais["tipo_veiculo"].value_counts()
display(fatalidades_por_veiculo.head(10))

# Gerar tabela formatada em Markdown
print("| Posição |   Tipo de Veículo   | Quantidade |")
print("|---------|---------------------|------------|")
for posicao, (veiculo, qtd) in enumerate(fatalidades_por_veiculo.head(10).items()):

    if posicao < 9:
        print(f"| 0{posicao + 1}º     | {veiculo.ljust(18)}  | {str(qtd).rjust(9)}  |")
    
    else:
        print(f"| {posicao + 1}º     | {veiculo.ljust(18)}  | {str(qtd).rjust(9)}  |")
```

```python
plt.figure(figsize=(12,6))
fatalidades_por_veiculo.head(10).sort_values(ascending=True).plot(kind = "barh", color = "red")
plt.title("Veículos com Maiores Registro de Acidentes com Fatalidades em Rodovias Federais")
plt.xlabel("Registros de Acidentes com Fatalidades")
plt.xticks(rotation = 0)
plt.savefig("fatalidades_por_veiculo")
plt.show()
```
